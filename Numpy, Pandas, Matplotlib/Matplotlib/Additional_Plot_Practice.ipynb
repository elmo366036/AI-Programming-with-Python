**Task 1**: Plot the distribution of combined fuel mileage (column 'comb', in miles per gallon) by manufacturer (column 'make'), for all manufacturers with at least eighty cars in the dataset. Consider which manufacturer order will convey the most information when constructing your final plot. **Hint**: Completing this exercise will take multiple steps! Add additional code cells as needed in order to achieve the goal.

```python
# get list of manufacturers greater than 80
greater_80 = fuel_econ.groupby('make').count()['id']>80
greater_80 = greater_80[greater_80==True]
greater_80_makes = greater_80.index.values

# get 'comb' data for these makes from fuel_econ, group and average them, and sort them in descending order
fuel_econ_80 = fuel_econ[fuel_econ['make'].isin(greater_80_makes)]
greater_80_makes_comb = fuel_econ_80[['make', 'comb']]
greater_80_makes_sorted = greater_80_makes_comb.groupby(['make']).mean().sort_values('comb', ascending=False).index

# create FacetGrid and plot data. Use min and max from greater_80_makes_comb.describe() for bin size
g = sb.FacetGrid(data=greater_80_makes_comb, col='make', col_wrap=6, col_order=greater_80_makes_sorted, size=2)
bins = np.arange(10, 60+2, 2)
g.map(plt.hist, 'comb', bins=bins)
#greater_80_makes_comb.describe()
```

**Task 2**: Continuing on from the previous task, plot the mean fuel efficiency for each manufacturer with at least 80 cars in the dataset.

```python
base_color = sb.color_palette()[0]
line_color = sb.color_palette('dark')[0]

# create df of makes (>80) with the comb mean and comb std. Order it by comb_mean
result = greater_80_makes_comb.groupby(['make']).mean()
result['comb_std'] = greater_80_makes_comb.groupby(['make']).std()
result.reset_index(inplace=True)
result.columns = ['make', 'comb_mean', 'comb_std']
result.sort_values('comb_mean', ascending=False, inplace=True)

# plot it with xerr bar
sb.barplot(data=result, x='comb_mean', y='make', color=base_color, xerr=result['comb_std'])
plt.xlabel('Average Combined Fuel Eff. (mpg)')
```
